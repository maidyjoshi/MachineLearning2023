Langkah 0 - Load Library

```python
import pandas as pd # import library pandas untuk membaca dan menulis data dari 
                    # berbagai sumber seperti CSV, Excel, SQL, dan lainnya ke dalam DataFrame

from sklearn.preprocessing import LabelEncoder, StandardScaler # mengimpor dua kelas dari library sklearn.preprocessing: LabelEncoder dan StandardScaler
                                                               # LabelEncoder adalah kelas yang digunakan untuk mengkodekan label atau nilai kategori dalam bentuk string atau teks menjadi bilangan bulat.
                                                               # StandardScaler adalah kelas yang digunakan untuk melakukan penskalaan (scaling) pada fitur-fitur numerik dalam dataset.
```

Langkah 1 - Load data

```python
dpath = 'Titanic-Dataset-fixed.csv' # mendefinisikan variabel dpath dan menginisialisasinya dengan string yang berisi path (lokasi) file CSV yang akan dibaca. Dalam kasus ini, file CSV yang akan dibaca adalah "Titanic-Dataset-fixed.csv".
df = pd.read_csv(dpath) #  menggunakan library Pandas untuk membaca data dari file CSV yang telah ditentukan dalam variabel dpath.
df.head() # melakukan pengecekan informasi tentang data
```

Langkah 2 - Slice Data

```python
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']] #  digunakan untuk memilih hanya beberapa kolom tertentu dari DataFrame df dan menyimpannya kembali dalam DataFrame yang sama dengan nama df
df.head() # melakukan pengecekan informasi tentang data
```

Langkah 3 - Encoding

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['Sex'] = le.fit_transform(df['Sex']) # proses dari encoding data frame 'sex'
df['Cabin'] = le.fit_transform(df['Cabin']) # proses dari encoding data frame 'cabin'
```

Langkah 4 - Verifikasi Hasil

```python
df.head()
```

Langkah 5 - Standarisasi

```python
std = StandardScaler() #  Ini adalah langkah pertama di mana Anda membuat objek StandardScaler dan menyimpannya dalam variabel std.
df['Age'] = std.fit_transform(df[['Age']]) #  pemilihan kolom Age dalam df tunggal yang menggunakan metode menghitung parameter yang telah diubah menjadi skala yang baru, di mana nilai rata-rata menjadi 0 dan deviasi standar menjadi 1. Data yang telah diubah ini kemudian disimpan kembali ke dalam kolom "Age" dalam DataFrame df, menggantikan data asli.
```
